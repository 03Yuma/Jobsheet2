```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

data = "wbc.csv"
df = pd.read_csv(data)
```

```python
df.info()
```

```python
df.head()
```

```python
#memisahkan variabel yang digunakan dan tidak digunakan

used_columns = ['id','diagnosis','radius_mean','area_mean']
df_used = df[used_columns]

unused_columns =df.columns.difference(used_columns)
df_unused = df[unused_columns]
```

```python
df_used.info()
```

```python
df_unused.info()
```

```python
#melakuakn encoding pada kolom diagnosis

new = LabelEncoder()
df_used['diagnosis']=new.fit_transform(df_used['diagnosis'])
```

```python
df_used.head()
```

```python
#melakukan standarisasi data

features = df.select_dtypes(include=['float','int64','int32'])
standar =StandardScaler()

df_scaled = df.copy()

df_scaled[features.columns] = standar.fit_transform(features)
```

```python
df_scaled.head()
```

```python
#melakukan split data

from sklearn.model_selection import train_test_split

# Stratified split dengan rasio 80:20, tetap menggunakan DataFrame 'df' langsung
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

# Menampilkan ukuran data setelah split
print(f'Jumlah data asli: {df.shape[0]}')
print(f'Jumlah data train: {df_train.shape[0]}')
print(f'Jumlah data test: {df_test.shape[0]}')
```
